```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read dataset
df_train = pd.read_csv("train.tsv", sep="\t", encoding="utf8")
df_test = pd.read_csv("test.tsv", sep="\t", encoding="utf8")
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
print(df_train.shape)
ax = sns.countplot(x='lable', data=df_train)
ax.set(title='Train set distribution')
ax.bar_label(ax.containers[0])
plt.show()

print(df_test.shape)
ax = sns.countplot(x='lable', data=df_test)
ax.set(title='Train set distribution')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
print("-------ALL-------")
print(df_train['AB'].apply(lambda x: len(x.split(" "))).describe())
for category in ['1a','1b','2a','2b','3a','3b','4']:
    print(f"-------{category}-------")
    print(df_train[df_train['lable']==category]['AB'].apply(lambda x: len(x.split(" "))).describe())
```

```python
print("-------ALL-------")
print(df_test['AB'].apply(lambda x: len(x.split(" "))).describe())
for category in ['1a','1b','2a','2b','3a','3b','4']:
    print(f"-------{category}-------")
    print(df_test[df_test['lable']==category]['AB'].apply(lambda x: len(x.split(" "))).describe())
```
